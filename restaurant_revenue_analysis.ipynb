# Выводы по обзору данных

## 1. Общая оценка качества данных

Данные в целом имеют высокое качество:

- пропуски отсутствуют

## 2. Структура данных

Грейн таблицы это блюдо внутри чека.

- один чек представлен несколькими строками
- значения, относящиеся к чеку (например, гости), дублируются

Для корректной аналитики требуется агрегация до уровня чека.

## 3. Особенности данных

В данных присутствуют позиции с нулевой стоимостью:

- модификаторы (добавки, типы сервировки и т.д.)
- сервисные позиции

Особенности:

- часть таких позиций не являются самостоятельными продажами (модификаторы с нулевой стоимостью)
- их стоимость уже включена в основное блюдо
- при этом в данных также присутствуют добавки с положительной стоимостью (гости доплачивают за них)

Следствие:

- позиции с нулевой стоимостью не учитываются при расчёте количества проданных блюд
- позиции с положительной стоимостью (в том числе добавки) учитываются как отдельные продажи
- нулевые позиции не влияют на выручку

Дополнительно:

- префикс "Д." в названии блюда или категории обозначает доставку, но не является полностью надёжным признаком (возможны ошибки справочника)

## 4. Несоответствие даты чека

Обнаружены чеки с несколькими значениями учетной даты:

- ~0.05% чеков (7 из ~12 800)

Детальный анализ показал:

- ранние даты содержат:
  - revenue = 0
  - qty = 0
  - guests = 0
- поздние даты содержат реальные значения продаж

Интерпретация:

- ранние записи являются техническими
- фактическая продажа отражена в более поздней дате

Принятое решение:

- для каждого чека используется максимальная дата

## 5. Несоответствие количества гостей чека

Обнаружены чеки с несколькими значениями количества гостей:

- ~0.5% чеков (62 из ~12 800)

Анализ показал:

- guests = 0 соответствует строкам с:
  - revenue = 0
  - qty = 0
- ненулевые значения гостей соответствуют реальным продажам

Интерпретация:

- нулевые значения — технические записи
- реальные значения гостей присутствуют в строках с продажами

Принятое решение:

- для каждого чека используется максимальное значение гостей

## 6. Отрицательные значения выручки

Обнаружены единичные отрицательные значения (~10 строк):

- связаны с операцией "Округление в пользу гостя"

Интерпретация:

- это корректировки, а не возвраты

Решение:

- учитывать в выручке без изменений

## 7. Особенности показателя "гости"

В данных присутствуют значения guests = 0.

Согласно описанию:

- для доставки количество гостей формируется на основании количества выбранных приборов, а не фактического числа людей

Детальный анализ показал:

- значения guests = 0 могут соответствовать:
  - техническим строкам (в проблемных чеках)
  - доставочным и нересторанным заказам, где гости фактически есть, но не указаны

Следствие:

- после исключения технических строк значение guests = 0 можно интерпретировать как отсутствие валидной информации о количестве гостей (в основном для доставки)
- показатель "гости" не может использоваться напрямую для всех типов заказов
- метрики, связанные с гостями, корректно рассчитывать только для ресторанных заказов

# Data Overview Conclusions

## 1. Overall Data Quality Assessment

The data overall has high quality:

- no missing values
- no duplicate rows
- data types are correctly defined (dates, numeric fields)

This allows proceeding to analysis without preliminary basic-level data cleaning.

## 2. Data Structure

The grain of the table is a dish within a check.

- one check is represented by multiple rows
- values related to the check (e.g., guests) are duplicated

For correct analysis, aggregation to the check level is required.

## 3. Data Specifics

The data contains positions with zero cost:

- modifiers (add-ons, serving types, etc.)
- service positions

Specifics:

- some of these positions are not standalone sales (zero-cost modifiers)
- their cost is already included in the main dish
- at the same time, the data also contains add-ons with positive cost (guests pay extra for them)

Implication:

- positions with zero cost are not included in the calculation of the number of dishes sold
- positions with positive cost (including add-ons) are treated as separate sales
- zero-cost positions do not affect revenue

Additionally:

- the prefix "Д." in dish or category names indicates delivery but is not fully reliable (reference data inconsistencies may exist)

## 4. Check Date Inconsistency

Checks with multiple accounting dates were identified:

- ~0.05% of checks (7 out of ~12,800)

Detailed analysis showed:

- earlier dates contain:
  - revenue = 0
  - qty = 0
  - guests = 0
- later dates contain actual sales values

Interpretation:

- earlier records are technical
- actual sales are reflected in the later date

Decision:

- the maximum date is used for each check

## 5. Check Guest Count Inconsistency

Checks with multiple guest values were identified:

- ~0.5% of checks (62 out of ~12,800)

Analysis showed:

- guests = 0 corresponds to rows with:
  - revenue = 0
  - qty = 0
- non-zero guest values correspond to actual sales

Interpretation:

- zero values are technical records
- actual guest values exist in rows with sales

Decision:

- the maximum guest value is used for each check

## 6. Negative Revenue Values

A small number of negative values were identified (~10 rows):

- associated with the operation "Rounding in favor of the guest"

Interpretation:

- these are adjustments, not returns

Decision:

- include them in revenue without changes

## 7. Specifics of the "guests" Metric

The data contains values where guests = 0.

According to the description:

- for delivery, the number of guests is derived from selected utensils, not actual people

Detailed analysis showed:

- guests = 0 may correspond to:
  - technical rows (in problematic checks)
  - delivery and non-restaurant orders where guests are not specified

Implication:

- after excluding technical rows, guests = 0 can be interpreted as missing valid guest information (primarily for delivery)
- the "guests" metric cannot be used directly across all order types
- guest-related metrics should only be calculated for restaurant orders

## 8. Sales Channel Structure

Order distribution:

- ~74% — restaurant orders
- ~26% — delivery, aggregators, and pickup

Conclusion:

- the core business is an offline restaurant
- delivery is an additional channel
- `order_type` is used as the most reliable source for channel classification

## Final Conclusion

The data is of high quality and suitable for analysis, but contains several specifics related to accounting logic:

- data is at the dish level
- technical rows are present (zero values)
- rare inconsistencies in dates and guest counts exist
- revenue adjustments are present
- the guest metric is not valid for part of the channels (delivery)

All identified specifics have been analyzed and accounted for, enabling correct aggregation and transition to business metric calculations.

# Подготовка витрины уровня чека (df_check)

На этапе подготовки данных выполнена нормализация исходных записей и приведение их к уровню чека с учетом особенностей учета в системе.

## 1. Особенности строк с revenue = 0

В данных присутствует значительное количество строк с нулевой выручкой (`revenue = 0`).

Анализ показал, что такие строки включают:

- модификаторы (например, "- молоко", "- с корицей")
- бесплатные дополнения (например, "Лимон 40г", "Кантуччи")
- служебные позиции (например, "Сервисный сбор", маркировка)
- реальные блюда, переданные бесплатно (акции, комплименты)

Таким образом, значение `revenue = 0` не является признаком ошибки или невалидной строки.

### Принятое решение

Строки с `revenue = 0` сохраняются в данных.

При этом:

- выручка агрегируется по всем строкам
- количество блюд (`qty`) и продуктовая аналитика рассчитываются только по строкам с `revenue > 0`

Это позволяет исключить технические позиции из продуктовых метрик без потери информации.

---

## 2. Обработка чеков с нулевой выручкой

После агрегации на уровень чека были выявлены чеки с `revenue = 0`:

- 315 чеков (~2.4% от общего числа)

Такие чеки могут отражать:

- полностью бесплатные заказы
- внутренние операции учета
- технические записи

### Принятое решение

Чеки с нулевой выручкой не удаляются.

Вводится флаг:

- `is_paid = revenue > 0`

### Использование

- выручка рассчитывается по всем чекам
- количество чеков может считаться:
  - по всем чекам
  - только по платным (`is_paid = True`)
- метрики, связанные с продажами (например, AOV), рассчитываются только по платным чекам

# Check-level data mart preparation (df_check)

At this stage, the raw data was normalized and transformed to the check level, taking into account the specifics of the accounting system.

## 1. Rows with revenue = 0

A significant number of rows with `revenue = 0` were identified.

Analysis showed that these rows include:

- modifiers (e.g., "- milk", "- cinnamon")
- free additions (e.g., "Lemon 40g", "Cantucci")
- service/technical positions (e.g., "Service fee", labeling)
- actual dishes provided for free (promotions, complimentary items)

Thus, `revenue = 0` does not indicate invalid or erroneous data.

### Decision

Rows with `revenue = 0` are retained.

At the same time:

- revenue is aggregated across all rows
- product-related metrics (`qty`, top dishes) are calculated only for rows with `revenue > 0`

This approach excludes technical positions from product analytics without losing data.

---

## 2. Checks with zero revenue

After aggregation to the check level, checks with `revenue = 0` were identified:

- 315 checks (~2.4% of total)

These may represent:

- fully free orders
- internal accounting operations
- technical records

### Decision

Zero-revenue checks are not removed.

A flag is introduced:

- `is_paid = revenue > 0`

### Usage

- revenue is calculated using all checks
- number of checks can be calculated:
  - either for all checks
  - or only for paid checks (`is_paid = True`)
- sales-related metrics (e.g., AOV) are calculated only for paid checks

---

## 3. Channel mapping

The `order_type` field contains raw system values and is not suitable for direct business analysis.

It was mapped into aggregated business channels:

- Restaurant — dine-in and veranda
- Delivery — in-house delivery
- Aggregator — third-party delivery platforms
- Pickup — customer pickup
- Banquet — banquet orders

Distribution:

- Restaurant — 8153 (~63%)
- Aggregator — 2731 (~21%)
- Delivery — 1465 (~11%)
- Pickup — 461 (~4%)
- Banquet — 74 (<1%)

Additionally:

- `is_delivery = channel IN (Delivery, Aggregator)`

Data was aggregated to the check level.

This layer removes row-level duplication and establishes "check" as the core business entity.

The next step is to build an aggregated daily-level data mart (`df_daily`).

Objective:

- move from transaction-level data to business metrics

The following metrics will be calculated:

- daily revenue
- number of checks

This mart aggregates data at the check level and enables basket and behavioral analysis.

### 1. Overall structure

- total checks: ~12.8K
- average check:
  - 2025: ~3655
  - 2026: ~3529
- median check:
  - 2025: ~2575
  - 2026: ~2609

Conclusion:
average check declines slightly, while median remains stable → drop driven by large checks

---

### 2. Basket size

- average items per check:
  - 2025: ~5.9
  - 2026: ~5.6
- median: 5
- most checks: 3–7 items

Conclusion:
typical basket contains 4–6 items

---

### 3. Distribution

- right-skewed distribution
- presence of large checks (up to 50 items)
- long tail

Conclusion:
there is a segment of large/group orders

---

### 4. Basket size vs revenue

Almost linear relationship:

- 1 item → ~750
- 5 items → ~2800
- 10 items → ~5600
- 20 items → ~13K+

Conclusion:
revenue is primarily driven by basket size

---

### 5. Basket diversity

- more categories → higher revenue:
  - 1 category → ~1100
  - 5 categories → ~3600
  - 10 categories → ~10K+

Conclusion:
diversity of basket strongly impacts revenue

---

### 6. Customer behavior

Two main patterns:

#### Small basket
- 1–3 items
- low revenue
- single category

#### Extended basket
- 5+ items
- multiple categories
- significantly higher revenue

---

### 7. Key business insight

Revenue growth is driven not by price but by:

- increasing number of items
- expanding basket composition

### 8. Business interpretation
