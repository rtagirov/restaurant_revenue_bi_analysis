```python
problem_checks = df_fact.groupby("check_id")["date"].nunique()
problem_checks = problem_checks[problem_checks > 1].index

df_fact[df_fact["check_id"].isin(problem_checks)].sort_values(["check_id", "date"])
```

```python
problem_date_checks = (
    df_fact
    .groupby("check_id")["date"]
    .nunique()
)

problem_date_checks = problem_date_checks[problem_date_checks > 1].index

date_check_summary = (
    df_fact[df_fact["check_id"].isin(problem_date_checks)]
    .groupby(["check_id", "date"])
    .agg(
        rows=("dish", "count"),
        revenue_sum=("revenue", "sum"),
        qty_sum=("qty", "sum"),
        guests_max=("guests", "max")
    )
    .reset_index()
    .sort_values(["check_id", "date"])
)

date_check_summary
```

```python
problem_guest_checks = (
    df_fact
    .groupby("check_id")["guests"]
    .nunique()
)

problem_guest_checks = problem_guest_checks[problem_guest_checks > 1].index

guest_check_summary = (
    df_fact[df_fact["check_id"].isin(problem_guest_checks)]
    .groupby(["check_id", "guests"])
    .agg(
        rows=("dish", "count"),
        revenue_sum=("revenue", "sum"),
        qty_sum=("qty", "sum")
    )
    .reset_index()
    .sort_values(["check_id", "guests"])
)

guest_check_summary
```

# Выводы по обзору данных

## 1. Общая оценка качества данных

Данные в целом имеют высокое качество:

- пропуски отсутствуют
- дубликаты строк отсутствуют
- типы данных корректно определены (даты, числовые поля)

Это позволяет перейти к аналитике без предварительной очистки базового уровня.

## 2. Структура данных

Грейн таблицы это блюдо внутри чека.

- один чек представлен несколькими строками
- значения, относящиеся к чеку (например, гости), дублируются

Для корректной аналитики требуется агрегация до уровня чека.

## 3. Особенности данных

В данных присутствуют позиции с нулевой стоимостью:

- модификаторы (добавки, типы сервировки и т.д.)
- сервисные позиции

Особенности:

- часть таких позиций не являются самостоятельными продажами (модификаторы с нулевой стоимостью)
- их стоимость уже включена в основное блюдо
- при этом в данных также присутствуют добавки с положительной стоимостью (гости доплачивают за них)

Следствие:

- позиции с нулевой стоимостью не учитываются при расчёте количества проданных блюд
- позиции с положительной стоимостью (в том числе добавки) учитываются как отдельные продажи
- нулевые позиции не влияют на выручку

Дополнительно:

- префикс "Д." в названии блюда или категории обозначает доставку, но не является полностью надёжным признаком (возможны ошибки справочника)

## 4. Несоответствие даты чека

Обнаружены чеки с несколькими значениями учетной даты:

- ~0.05% чеков (7 из ~12 800)

Детальный анализ показал:

- ранние даты содержат:
  - revenue = 0
  - qty = 0
  - guests = 0
- поздние даты содержат реальные значения продаж

Интерпретация:

- ранние записи являются техническими
- фактическая продажа отражена в более поздней дате

Принятое решение:

- для каждого чека используется максимальная дата

## 5. Несоответствие количества гостей чека

Обнаружены чеки с несколькими значениями количества гостей:

- ~0.5% чеков (62 из ~12 800)

Анализ показал:

- guests = 0 соответствует строкам с:
  - revenue = 0
  - qty = 0
- ненулевые значения гостей соответствуют реальным продажам

Интерпретация:

- нулевые значения — технические записи
- реальные значения гостей присутствуют в строках с продажами

Принятое решение:

- для каждого чека используется максимальное значение гостей

## 6. Отрицательные значения выручки

Обнаружены единичные отрицательные значения (~10 строк):

- связаны с операцией "Округление в пользу гостя"

Интерпретация:

- это корректировки, а не возвраты

Решение:

- учитывать в выручке без изменений

## 7. Особенности показателя "гости"

В данных присутствуют значения guests = 0.

Согласно описанию:

- для доставки количество гостей формируется на основании количества выбранных приборов, а не фактического числа людей

Детальный анализ показал:

- значения guests = 0 могут соответствовать:
  - техническим строкам (в проблемных чеках)
  - доставочным и нересторанным заказам, где гости фактически есть, но не указаны

Следствие:

- после исключения технических строк значение guests = 0 можно интерпретировать как отсутствие валидной информации о количестве гостей (в основном для доставки)
- показатель "гости" не может использоваться напрямую для всех типов заказов
- метрики, связанные с гостями, корректно рассчитывать только для ресторанных заказов

## 8. Структура каналов продаж

Распределение заказов:

- ~74% — заказы в ресторане
- ~26% — доставка, агрегаторы и самовывоз

Вывод:

- основной бизнес это офлайн-ресторан
- доставка является дополнительным каналом
- для классификации каналов используется поле order_type как наиболее надёжный источник

## Следующий шаг

На основе проведённого анализа выполняется переход к формированию витрины уровня чека (`df_check`).

Цель:

- перейти от уровня строк (блюд) к уровню бизнес-сущности (чек)

В рамках следующего шага:

- данные агрегируются по `check_id`
- применяется логика:
  - `date = max(date)`
  - `guests = max(guests)`
  - `revenue = sum(revenue)`
- сохраняются все чеки, включая чеки с нулевой выручкой
- вводятся дополнительные признаки:
  - `is_paid` — флаг платного чека
  - `channel` — укрупнённый канал продаж
  - `is_delivery` — флаг доставки

Результатом является витрина `df_check`, которая используется как базовый слой для расчёта метрик и построения последующих агрегированных витрин.

# Data Overview Conclusions

## 1. Overall Data Quality Assessment

The data overall has high quality:

- no missing values
- no duplicate rows
- data types are correctly defined (dates, numeric fields)

This allows proceeding to analysis without preliminary basic-level data cleaning.

## 2. Data Structure

The grain of the table is a dish within a check.

- one check is represented by multiple rows
- values related to the check (e.g., guests) are duplicated

For correct analysis, aggregation to the check level is required.

## 3. Data Specifics

The data contains positions with zero cost:

- modifiers (add-ons, serving types, etc.)
- service positions

Specifics:

- some of these positions are not standalone sales (zero-cost modifiers)
- their cost is already included in the main dish
- at the same time, the data also contains add-ons with positive cost (guests pay extra for them)

Implication:

- positions with zero cost are not included in the calculation of the number of dishes sold
- positions with positive cost (including add-ons) are treated as separate sales
- zero-cost positions do not affect revenue

Additionally:

- the prefix "Д." in dish or category names indicates delivery but is not fully reliable (reference data inconsistencies may exist)

## 4. Check Date Inconsistency

- actual sales are reflected in the later date

## 5. Check Guest Count Inconsistency

- ~0.5% of checks (62 out of ~12,800)

  - revenue = 0
  - qty = 0

Interpretation:

- zero values are technical records
- actual guest values exist in rows with sales

Decision:

- the maximum guest value is used for each check

## 6. Negative Revenue Values

A small number of negative values were identified (~10 rows):

- associated with the operation "Rounding in favor of the guest"

Interpretation:

- these are adjustments, not returns

Decision:

- include them in revenue without changes

## 7. Specifics of the "guests" Metric

The data contains values where guests = 0.

According to the description:

- for delivery, the number of guests is derived from selected utensils, not actual people

Detailed analysis showed:

- after excluding technical rows, guests = 0 can be interpreted as missing valid guest information (primarily for delivery)

- guest-related metrics should only be calculated for restaurant orders
